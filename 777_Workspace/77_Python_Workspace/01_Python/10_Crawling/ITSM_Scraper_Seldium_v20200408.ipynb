```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
import os
import json
import requests
import time
from bs4 import BeautifulSoup

import pandas as pd
```

```python
LOGIN_URL = 'http://itsm.ktds.co.kr/kt/index.do' 
SERVICE_OPERATION_STATUS = 'http://itsm.ktds.co.kr/kt/servicedesk/list.do?selectedMnuID=USR00041&mnuIdx=0&brdOid=0&searchReqClass=COD_01694'
```

```python
def get_driver():
    driver = webdriver.Chrome(r'C:\Python_Lib\chromedriver.exe')
    driver.implicitly_wait(3)
 
    return driver
```

```python
def get_config():
    try:
        with open('config.json') as json_file:
            json_data = json.load(json_file)
    except Exception as e:
        print('Error in reading config file, {}'.format(e))
        return None
    else:
        return json_data
```

```python
def login_itsm(driver, id, pw):
    element = driver.find_element_by_id("LoginId")
    element.send_keys(id)
    
    element = driver.find_element_by_id("LoginPw")
    element.send_keys(pw)

    element = WebDriverWait(driver, 10).until(
        EC.presence_of_element_located((By.CSS_SELECTOR, "input.submit"))
    )
    element.click()
 
    return False

```

```python
def get_srm_detailed_infromation(driver, SRM_number_list):
    for line in SRM_number_list:
        srm = line
        print(srm)
        r = driver.get(SERVICE_OPERATION_STATUS)
        element = driver.find_element_by_id("SearchIncidentID")
        element.send_keys(srm)

        time.sleep(1) 
        element = driver.find_element_by_xpath('//*[@id="SearchBtn"]/img')
        element.click()

        time.sleep(1) 
        element = driver.find_element_by_class_name('datagrid-row')
        r= element.click()

        soup = BeautifulSoup(driver.page_source, 'html')
        soup_text = soup.find("th", text="1선 조치 내역").find_next_sibling("td").text
        soup_text = soup_text.replace("\n","")
        soup_text = soup_text.lstrip()
        print(soup_text)
        resolution_list.append(soup_text)
#         time.sleep(1)

    return False

```

```python
resolution_list = list()

driver = get_driver()
driver.get(LOGIN_URL)
config = get_config()
print(config['userId'], config['userPw'])
login_itsm(driver, config['userId'], config['userPw'])

SRM_INFO = pd.read_excel("C:/Users/neo/Downloads/서비스문의 목록 (3).xls", index_col=None, header = 1)
SRM_Number_List = SRM_INFO["서비스문의 ID"]
SRM_Number_List_Sample = SRM_Number_List[:]
time.sleep(1)
get_srm_detailed_infromation(driver, SRM_Number_List_Sample)
```
